```python
emotions = {
    0: 'Angry', 
    1: 'Disgust', 
    2: 'Fear', 
    3: 'Happy', 
    4: 'Sad', 
    5: 'Surprise', 
    6: 'Neutral'
}
```

```python
def img_visualizer(df, row_index, rows=1, cols=1):
    
    if type(row_index) != list:
        row_index = [row_index]
    assert len(row_index) == rows*cols,"number row index list must be equal to number of rows time number of columns"

    emotions = {
        0: 'Angry', 
        1: 'Disgust', 
        2: 'Fear', 
        3: 'Happy', 
        4: 'Sad', 
        5: 'Surprise', 
        6: 'Neutral'}
    
    n = 0
    fig, axs = plt.subplots(rows, cols)
    for i in range(1,rows*cols+1):
            pixels = df["pixels"][row_index[n]]
            img = np.array(pixels.split(' '),dtype='int').reshape((48,48))
            plt.subplot(rows, cols, i)
            plt.imshow(img,cmap='gray')
            plt.axis('off')
            plt.title(emotions[df["emotion"][row_index[n]]])
            n+=1

    
```

```python
img_visualizer(fer_df, [i*100 for i in range(16)],4,4)
```
